```python
import os
import matplotlib.pyplot as plt
import pandas as pd

from utils.Pyutils import load_pkl

plt.style.use('bmh')
plt.rcParams['lines.linewidth'] = 1
```

```python
model_names = ["transformer_encoder_mlp_decoder", "transformer_encoder_mlp_decoder2"]
```

```python
accs = []
for model in model_names:
    output_path = os.path.join(os.getcwd(),
                            "results",
                            model)

    data = load_pkl(os.path.join(output_path, "results.pickle"))
    accs.append(data["acc"].rename(columns={"acc": model}))
accs_df = pd.concat(accs, axis=1)
```

```python
accs_df.plot(figsize=(10, 6))
```
